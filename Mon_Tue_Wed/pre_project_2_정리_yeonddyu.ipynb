# 분석 요약

## 분석 목적
- 현장이슈  
: 콘택트렌즈의 도수에 영향을 주는 요인이 다양하고 작업자의 숙련도에 따라 금형조합 별 도수 적중률이 크게 변화하는 문제가 있다.
<br><br>
- 목적  
: 콘택트렌즈의 도수와 금형조합 간의 상관관계를 분석을 통해 콘택트렌즈의 도수를 예측하고 최적화 알고리즘을 이용하여 최적의 금형조합을 구한다.

## 콘택트렌즈 생산과정
: 콘택트렌즈는 렌즈의 양면 형태를 가진 렌즈의 상형, 하형 몰드를 만들고, 원재료를 하형 몰드에 주입한 후 상형 몰드를 올려 렌즈를 성형하는 몰드캐스팅 방식으로 생산된다.  
상형, 하형 몰드는 렌즈의 곡률과 두께를 결정하는 주요 요인으로, 렌즈의 품질(도수)에 큰 영향을 미친다.

## 이슈사항
- 공정(설비)상의 문제 현황  
: 플라스틱 사출공정에서 만들어지는 상형, 하형 몰드는 렌즈의 양면 형태를 가지고 있 어, 콘택트렌즈의 내측, 외측 성형에 직접적인 영향을 미친다. 사출공정을 통해 생산된 상형, 하형 몰드 사이에 주입한 액상원료를 고온으로 고체화시키는 과정에서 플라스틱 몰드가 변형되면서 지시한 규격과 다른 규격의 제품이 생상되는 문제가 발생한다.  
: 현장 작업자는 설비, 설비 내 금형 설치 위치, 온도 등의 조건에 따른 플라스틱 몰드의 변형을 고려하여 생산할 제품 규격에 맞는 금형 조합을 준비해야 한다.  
: 숙련된 작업자는 축적해온 경험에 의존하여 금형 조합을 선별하지만, 작업자 숙련도에 따라 지시한 제품 규격과 생산된 제품의 규격 차이는 매우 크게 나타 날 수 있다.
<br><br>
- 문제해결 장애요인  
: 사출 공정에서 한 번의 공정 운영 시 8개의 몰드를 사출하는데, 8개 위치에서 동일한 제품군을 생산하는 경우는 거의 없고 각기 다른 규격의 제품군을 생산한다.  
: 플라스틱 몰드의 변형은 다양한 요인이 서로 복잡하게 영향을 미칠 수 있기 때문에 액상원료 고체화 시 가해지는 열의 온도, 설비 내 금형의 위치 등 공정 조건 뿐만아니라 현장의 다양한 환경적인 조건도 고려해야 한다.
<br><br>
- 극복 방안  
: 현장에서 확보할 수 있는 데이터를 기반으로 렌즈의 도수에 영향을 미치는 요인 을 탐색하고 변수 간 관계를 파악한다. 금형 조합 탐 색 후 설비, 금형 위치를 고려하여 도수를 예측하고, 생산하고자 하는 목표 도수와 비 교하여 차이가 작은 순으로 5개의 금형 조합을 추천한다.

## 누락 데이터 처리 함수 정의

```python
## null값이거나 공백인 데이터 값을 np.NaN으로 수정 후
## CORES, RADIUS 중 np.NaN 값을 가지는 행 제거
def preprocessing_missing(df_before):
    type = 'Missing Value'
    print(f'Type of preprocessing: {type}')
    print(f'The shape of data before preprocessing: {df_before.shape}')
    
    tmp = np.where((df_before.values == 'NULL')|(df_before.values == ''), np.NaN, df_before.values)
    df_after = pd.DataFrame(data=tmp, columns=df_before.columns)
    df_after = df_after.dropna(how='any', subset=CORES+RADIUS)
    
    print(f'The shape of data after preprocessing: {df_after.shape}')
    
    return df_after
```

## 데이터 형변환 처리 함수 정의

```python
def preprocessing_conversion(df_before):
    type = 'Data Type Conversion'
    print(f'Type of preprocessing: {type}')
    print(f'The shape of data before preprocessing: {df_before.shape}')
    # 컬럼명을 모두 대문자로 변환
    df_before.columns = [col.upper() for col in df_before.columns]
    print(df_before.columns)
    
    # 컬럼의 데이터형 통일
    col_type = {'MOLD_POS':'int32',
                'REAL_POWER':'float32',
                'IN_RADIUS':'float32',
                'OUT_RADIUS':'float32',
                'POWER':'float32',
                'POWER1':'float32',
                'POWER2':'float32',
                'POWER3':'float32',
                'POWER4':'float32',
                'POWER5':'float32'}
                
    df_after = df_before.astype(col_type)
    df_after['MFG_DT'] = pd.to_datetime(df_before['MFG_DT'])

    # RADIUS 값의 부호 제거
    df_after['IN_RADIUS'] = abs(df_after['IN_RADIUS'])
    df_after['OUT_RADIUS'] = abs(df_after['OUT_RADIUS'])
    
    print(f'The shape of data preprocessing: {df_after.shape}')
    
    return df_after
```

## 지정 날짜 기준 데이터 추출 함수 정의

```python
## 모델의 정확성을 높이기 위해 2020년 1월 1일 이후 데이터만 분석
def preprocessing_date(df_before):
    type = 'Removing Irrelevant Data in terms of Date'
    print(f'Type of preprocessing: {type}')
    print(f'The shape of data before preprocessing: {df_before.shape}')
    
    df_after = df_before[df_before['MFG_DT'] >= np.datetime64('2020-01-01')]
    print(f'The shape of data preprocessing: {df_after.shape}')
    
    return df_after
```

## 이상치 처리 함수 정의

```python
def preprocessing_outlier(df_before):
    type = 'Removing Outlier'
    print(f'Type of preprocessing: {type}')
    print(f'The shape of data before preprocessing: {df_before.shape}')
    
    # 조건1: REAL_POWER와 POWER 값이 같은 경우
    mask1 = (df_before['REAL_POWER'] == df_before['POWER'])
    # 조건2: POWER1, POWER2, POWER3, POWER4, POWER5의 평균과 POWER의 오차가 0.5보다 작은 경우
    mask2 = abs(mean(df_before[POWERS], axis=1) - df_before['POWER']) < 0.5
    # 조건1과 조건2를 모두 만족하는 경우만 추출
    df_after = df_before[mask1 & mask2]
    
    
    # IsolationForest 적용 전의 데이터 시각화
    fig = plt.figure(figsize=(16, 16))
    ax_before = fig.add_subplot(1, 1, 1)
    df_after.plot.scatter(x='IN_RADIUS', y='OUT_RADIUS', ax=ax_before)
    
    # IsolationForest 적용 후의 데이터 시각화
    ifo = IsolationForest(random_state=0)
    results = ifo.fit_predict(df_after[RADIUS])
    df_after = df_after[results == 1]    # 1: 정상치, -1: 이상치
    df_after.plot.scatter(x='IN_RADIUS', y='OUT_RADIUS',
                          ax=ax_before, color='black')
    plt.show()
    
    print(f'The shape of data preprocessing: {df_after.shape}')
    
    return df_after
```

=>  Isolation Forest는 무작위로 데이터를 분할하여 이상치를 탐지하는 데 효과적이다. 고차원 공간에서도 잘 동작하며, 특히 데이터가 매우 차원이 높고 복잡한 경우에 유용하다.

# XGBoost 모델 학습

## 주요 변수 및 함수 정의

```python
## 원-핫 인코딩
ENCODER = None

## XGBoost
N_ESTIMATORS = 6000
ETA = 0.25
SUB_SAMPLE = 0.9
```

=> ENCODER는 XGBoost 학습과 예측에 동일한 컬럼을 가진 X 데이터를 생성하기 위해 사용한다.  
=> N_ESTIMATORS 와 같이 상수를 사용하는 XGBoost 모델 파라미터도 변수로 정의 한다.

```python
## 학습 및 평가 데이터셋 분리
def prepare_data(df, numeric_cols, onehot_cols, y_col):
    x_cols = numeric_cols + onehot_cols
    # 결측치 제거
    df = df.dropna(axis=0, how='any',
                   subset=x_cols+y_col)
    data_X = df[x_cols]
    data_y = df[y_col]
    
    print('\nSplitting data into train set and test set')
    X_train, X_test, y_train, y_test = train_test_split(data_X, data_y, test_size=0.3, random_state=42)
    
    print(f'The shape of X_train, y_train: {X_train.shape, y_train.shape}')
    print(f'The shape of X_test, y_test: {X_test.shape, y_test.shape}')
    
    # 수치형 데이터의 최솟값, 최댓값 (도수로 그룹)
    bounds = df.groupby(['REAL_POWER'])[numeric_cols].agg(['min', 'max'])
    
    return X_train, X_test, y_train, y_test, bounds
```

=> prepare_data는 모델 학습에 필요한 학습 데이터셋과 검증 데이터셋을 생성한다.

```python
## 원-핫 인코딩
def transform_cols(df, encoder):
    # encoder가 None이면 OneHotEncoder를 생성 후 학습, 반환
    if encoder is None:
        print('ENCODER fit_transform')
        encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
        onehot = encoder.fit_transform(df[EQUIP_INFO])
    # 모델 재사용
    else:
        print('ENCODER transform')
        onehot = encoder.transform(df[EQUIP_INFO])
        
    transformed_onehot_cols = encoder.get_feature_names_out(EQUIP_INFO).ravel().tolist()  # 다차원 배열을 1차원으로 변환
    df = np.hstack((df[RADIUS].to_numpy(), onehot))    # 주어진 배열들을 수평으로 쌓아 하나의 배열로 결합
    df = pd.DataFrame(df, columns=RADIUS+transformed_onehot_cols)
    
    return df, encoder
```

=> transform_cols 는 데이터셋의 컬럼을 원-핫 인코딩 형태로 변환한다.

# 연관규칙 결과와 XGBoost 예측 결과에서 최적조건 찾기

## 예측도수 변환에 필요한 변수 및 함수 정의

: POWER_VALUES는 0부터 –6 까지는 0.25 단위로 작아지고, -6부터 –10까지는 0.5 단위로 작아진다.  
: POWER_VALUES의 값 중에서 측정도수의 평균(POWER1 ~ POWER5)과 오차가 가장 작은 값이 최종도수로 결정된다.

```python
# POWER 할당 함수
POWER_VALUES = np.array([-0.25, -0.5, -0.75, -1.0, -1.25, -1.5, -1.75,-2.0, -2.25, -2.5, -2.75,
                         -3.0, -3.25, -3.5, -3.75, -4.0, -4.25, -4.5, -4.75, -5.0, -5.25, -5.5, -5.75,
                         -6.0, -6.5, -7.0, -7.5, -8.0, -8.5, -9.0, -9.5, -10.0])

def power_allocate(value):
    return POWER_VALUES[np.argmin(abs(POWER_VALUES - value))]

def vectorize(value):
    return np.vectorize(power_allocate)(value)
```

```python
# XGBoost Regressor 모델을 이용한 REAL_POWER 예측
df_result['PREDICTED_POWER'] = regressor.predict(X)
df_result['PREDICTED_POWER2'] = df_result['PREDICTED_POWER'].apply(vectorize)
```

## XGBoost 예측결과 내 우선순위 도출

```python
## 우선순위가 높은 5개의 최적조건 도출
# 예측 도수와 목표 도수의 오차를 RANK_INDEX_2에 저장
df_result['RANK_INDEX_2'] = np.abs(df_result['PREDICTED_POWER'].values - target_power)

# 입력조건과 일치하는 데이터가 있는 경우
# 최종 우선순위(RANK_INDEX는 RANK_INDEX_1과 RANK_INDEX_2의 합
if all(df_result['EQUIP_ID'] == equipment_id) & all(df_result['MOLD_POS'] == mold_position):
    df_result['RANK_INDEX'] = df_result['RANK_INDEX_1'] + df_result['RANK_INDEX_2']

# 일치하는 조건이 없는 경우
# RANK_INDEX_2를 최종 우선순위
else:
    df_result['RANK_INDEX'] = df_result['RANK_INDEX_2']
# 최종 우선순위가 높은 순으로 최대 5개의 최적조건을 도출
df_result = df_result.sort_values(by='RANK_INDEX').iloc[:5]

df_result['EQUIP_ID'] = equipment_id
df_result['MOLD_POS'] = mold_position
```

## 결과 출력

```python
## 결과테이블에 오차값 추가
# 사용자가 지정한 목표도수와 예측한 결정도수의 오차를 ERROR에 저장
df_result['ERROR'] = df_result['PREDICTED_POWER2'] - target_power
```
