# 지역별 범죄 발생 현황 분석

```python
df = pd.read_csv('./범죄_data/범죄_발생지_20240116161110.csv', header = 1)

theft = df.iloc[11] # 절도 범죄
df = df.drop('죄종별(1)', axis = 1)
df = df.drop('계', axis = 1).T
df = df.rename(columns = df.iloc[0])
df = df.drop(df.index[0])
df = df.drop('소계', axis = 1)
df = df.drop('죄종별(2)', axis = 1)


df
```

```python
theft_df = pd.DataFrame(theft)
theft_df = theft_df.iloc[3:]
theft_df = theft_df.rename(columns = {11 : '절도'})


df = pd.concat([df, theft_df], axis = 1)

df
```

```python
all_df = df.loc['서울' : '세종']

all_df
```

```python
df.index
```

```python
re_lst = ['경기도', '강원도', '충청북도', '충청남도', '경상북도', '경상남도', '제주특별자치도']

def si_re_change(df):
    ok_df = pd.DataFrame()
    for name in re_lst:
        re_df = df[df.index.str.contains(name)]
        re_df = re_df.replace('-', 0)
        re_df = re_df.astype('int')
        re_df = re_df.sum()
        re_df =  pd.DataFrame(re_df, columns = [name]).T
        ok_df = pd.concat([ok_df, re_df])
    return ok_df

ok_df = si_re_change(df)

# 전라남도, 전라북도는 이상해서 따로
Jeollanam = df[df.index.str.contains('전라남도')]
Jeollanam = Jeollanam.replace('-', 0)
Jeollanam = Jeollanam.astype('int')
Jeollanam = Jeollanam.sum()
Jeollanam = pd.DataFrame(Jeollanam, columns = ['전라남도']).T

Jeollabuk = df[df.index.str.contains('전라북도')]
Jeollabuk = Jeollabuk.replace('-', 0)
Jeollabuk = Jeollabuk.astype('int')
Jeollabuk = Jeollabuk.sum()
Jeollabuk = pd.DataFrame(Jeollabuk, columns = ['전라북도']).T


all_df = pd.concat([all_df, ok_df, Jeollanam, Jeollabuk])
all_df
```
